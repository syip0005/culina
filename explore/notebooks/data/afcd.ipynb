```python
df_nutrients = pd.read_excel(
    '../../data/afcd/AFCD Release 3 - Nutrient profiles.xlsx',
    sheet_name = 'All solids & liquids per 100 g',
    header=2
)
```

```python
df_nutrients.head()
```

```python
import csv

cols = df_nutrients.columns

file_path = '../../data/afcd/nutrient_cols.csv'

with open(file_path, mode='w', newline='', encoding='utf-8') as file:
    csv_writer = csv.writer(file)

    for item in cols:
        cleaned = " ".join(str(item).splitlines())   # removes \n, \r\n etc
        csv_writer.writerow([cleaned])

print(f"Data saved to {file_path}")
```

Nutrients appears to be more useful, and we can find the relevant columns required from that.

```python
df_nutrients['energy_kj'] = df_nutrients['Energy with dietary fibre, equated \n(kJ)']
df_nutrients['protein_g'] = df_nutrients['Protein \n(g)']
df_nutrients['fat_g'] = df_nutrients['Fat, total \n(g)']
df_nutrients['carbs_g'] = df_nutrients["Available carbohydrate, without sugar alcohols \n(g)"]

df_nutrients_compact = df_nutrients[[
    'Food Name',
    'energy_kj',
    'protein_g',
    'fat_g',
    'carbs_g'
]]
```
